### 2. 분산: 1/n * (실제값-평균)^2

```python
sumation = 0
meanValue = np.mean(sampleData1)
for eachItem in sampleData1:
    sumation = sumation + ( meanValue - eachItem ) ** 2
sampleData1Var = 1/datalen * sum
sampleData1Var
```

```python
def varCalc(inList):
    sumation = 0
    meanValue = np.mean(inList)
    for eachItem in inList:
        sumation = sumation + ( meanValue - eachItem ) ** 2
    outVar = 1/(datalen) * sumation
    return outVar
varCalc(sampleData2)
```

```python
print(  varCalc(sampleData1) ,
        varCalc(sampleData2) ,
     )
```

```python
import numpy as np
import pandas as pd

csData = pd.read_csv("../dataset/customerdata.csv") 
# 무이자 할부 2건 사용한 고객데이터의 평균구매가격정보를 리스트로 변환
boxData = list(csData.loc[ (csData.EMI==2) ].AVGPRICE) 

# 사분위수(Quartile), 사분범위(IQR) 계산
q1 = np.percentile(boxData, 25) #Q1 (first quartile) 
q2 = np.percentile(boxData, 50) #Q2 (median)
q3 = np.percentile(boxData, 75) #Q3 (third quartile)
iqr =(q3-q1)
print("Q1={} Q2={} Q3={} IQR={}".format(q1, q2, q3, iqr))

# 하한선, 상한선 계산
minimum = q1 - (iqr * 1.5) 
maximum = q3 + (iqr * 1.5) 
print("하한선={} 상한선={}".format(minimum, maximum))

# 데이터의 최소, 최대값 계산
minValue = min(boxData)
maxValue = max(boxData)  
print("데이터 최소값={} 데이터 최대값={}".format(minValue, maxValue))
```
