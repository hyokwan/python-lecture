### 1. 파일 데이터 불러오기

```python
# CSV 파일을 읽어 Data Frame 변수에 저장하기
selloutDataLocalUrl = "../dataset/kopo_product_volume.csv"
selloutDataGitUrl = "https://raw.githubusercontent.com/hyokwan/python-lecture/master/dataset/kopo_product_volume.csv"
selloutData = pd.read_csv(selloutDataGitUrl)
selloutData.head()
```

### 2. MySQL에 데이터프레임 저장

```python
# MySQL 데이터베이스 연결 정보 설정
user = 'kopo'
password = 'kopo'
host = '127.0.0.1'
port = '3306'
database = 'kopodb'

# MySQL 데이터베이스 연결 정보 설정 (클라우드)
user = 'hkcode'
password = 'hkcodemariadb'
host = '139.150.80.119'
port = '3306'
database = 'hkcodedb'

# SQLAlchemy 엔진 생성
engine = create_engine(f'mysql+mysqlconnector://{user}:{password}@{host}:{port}/{database}')

# 데이터프레임을 MySQL 데이터베이스의 테이블에 저장
table_name = 'kopo_product_volume_20240726'.upper()
selloutData.to_sql(name=table_name, con=engine, if_exists='replace', index=False)
```

```python
# MySQL 데이터베이스 연결 정보 설정 (ngrok)
user = 'kopo'
password = 'kopo'
host = '0.tcp.jp.ngrok.io'
port = '11169'
database = 'kodb'


# SQLAlchemy 엔진 생성
engine = create_engine(f'mysql+mysqlconnector://{user}:{password}@{host}:{port}/{database}')

# 데이터프레임을 MySQL 데이터베이스의 테이블에 저장
table_name = 'kopo_product_volume_ngrok'
selloutData.to_sql(name=table_name, con=engine, if_exists='replace', index=False)
```

### 3. MySQL 내 데이터 조회

```python
# 테이블 목록 조회
inspector = inspect(engine)

tables = inspector.get_table_names()
tables

# 타겟 테이블 조회
sqlSelect = """
SELECT *
FROM kopo_product_volume"""

selloutDf = pd.read_sql_query(sqlSelect, con=engine)
selloutDf.head()
```
