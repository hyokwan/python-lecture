```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine 
 
# CSV 파일을 읽어 Data Frame 변수에 저장하기
customerData = pd.read_csv("../dataset/customerdata.csv")

customerData.columns = ["CUSTID","AVGPRICE","EMI","DEVICECOUNT","PRODUCTAGE","CUSTTYPE"]
#customerData.columns = [x.upper() for x in customerData.columns]

# 데이터 VIEW
customerData.head()


customerData["PRODUCTAGE"]= np.\
where(customerData["PRODUCTAGE"] < 1, 1,
customerData["PRODUCTAGE"])

customerData.head()
```

```python
import numpy as np
customerData["PRODUCTAGE_NEW"] = np.where(customerData["PRODUCTAGE"] < 1, 1,\
                                          np.where(customerData["PRODUCTAGE"]<2, 2,\
                                          np.where(customerData["PRODUCTAGE"]<3, 3,\
                                          5)))
customerData.head()
```

### 다른문제 풀이방식

```python
customerData.loc[ (customerData.PRODUCTAGE < 1), "TEST"   ] = 1
customerData.loc[ (customerData.PRODUCTAGE >= 1) & (customerData.PRODUCTAGE < 2), "TEST"   ] = 2
customerData.loc[ (customerData.PRODUCTAGE >= 2) & (customerData.PRODUCTAGE < 3), "TEST"   ] = 3
customerData.loc[ (customerData.PRODUCTAGE >= 3) , "TEST"   ] = 5
```

```python
def refining(df):
    if df['PRODUCTAGE'] < 1:
        return 1
    elif df['PRODUCTAGE'] < 2:
        return 2
    elif df['PRODUCTAGE'] < 3:
        return 3
    else:
        return 5
```

```python
customerData["PRODUCTAGE_NEW2"]=customerData.apply(refining, axis=1)
```

# B. 데이터 합치기

```python
import pandas as pd
from sqlalchemy import create_engine 
 
# CSV 파일을 읽어 Data Frame 변수에 저장하기
selectExample = pd.read_csv("../dataset/kopo_channel_seasonality_new.csv")
print(selectExample.head())

# CSV 파일을 읽어 Data Frame 변수에 저장하기
regionMasterData = pd.read_csv("../dataset/kopo_region_mst.csv", encoding = "ms949") 

print(regionMasterData.head())
```

```python
mergeResult = pd.merge(selectExample, regionMasterData,\
        left_on="REGIONID", right_on = "REGIONID", how="left")\
[["REGIONID","REGIONNAME","PRODUCT","YEARWEEK","QTY"]]
mergeResult.head()
```

```python
len(mergeResult)
```

```python
import numpy as np

# A01 데이터 정제
a01data = selectExample[(selectExample['REGIONID']=='A01')]
print(len(a01data))

# A71 데이터 정제
a71data = selectExample[(selectExample['REGIONID']=='A02')]
a71data['QTY'] = np.where(a71data['QTY'] < 500,500,a71data['QTY'])
print(len(a71data))

# 데이터 결과 값 MERGE
mergedata = pd.concat([a01data, a71data], axis=0)
print(len(mergedata))
```
