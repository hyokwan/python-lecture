```python
import pandas as pd
# metaData = pd.read_excel("../dataset/indata/meta_data.xlsx")
# metaData

from glob import glob
import os
import csv

### 출력데이터 설정 (예: 구동코드 기준 상위폴더 내 dataset 폴더 내 outdata 폴더에 csv 대상자료가 있음
root = "../"
subdir = "outdata"

## *.csv 패턴으로 데이터 탐색
searchPattern = "*.csv"
targetPattern = os.path.join(root, subdir, searchPattern)
targetDf = glob(targetPattern)

## 결과리포트용 빈 데이터프레임 생성
empty_df = pd.DataFrame()

## 파일을 하나씩 탐색
fileLen = len(targetDf)
for i in range(0, fileLen):
    
    # 결과 첫번째 컬럼 : 파일명 정의
    fileName = targetDf[i]
    targetData = pd.read_csv(targetDf[i])
    # targetData = pd.read_csv(targetDataFrame, quoting = csv.QUOTE_NONE, delimiter = "\t")
    
    # 결과 두번째 컬럼 : 전체 행개수 정의
    totalRow = targetData.shape[0]   
    # 결과 세번째 컬럼 : 중복 행개수 정의
    duplRow = targetData.drop_duplicates().shape[0]   

    # 결과에 따라 PASS, FAIL 정의
    if(totalRow == duplRow):
        result="PASS"
    else:
        result="FAIL"
     
    finalData = pd.DataFrame(zip([fileName], 
                                 [totalRow], 
                                 [duplRow], 
                                 [result]), 
                 columns = ["파일명", "전체row개수", "전체중복row개수", "검증결과"])
    empty_df = empty_df.append(finalData)
    
empty_df.reset_index(drop=True)
```
