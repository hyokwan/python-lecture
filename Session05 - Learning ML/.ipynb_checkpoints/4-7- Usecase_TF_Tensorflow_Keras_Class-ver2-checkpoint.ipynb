```python
# 데이터 불러오기 및 정렬
featureData = pd.read_csv('../dataset/customerdata.csv')
featureData.head()
```

```python
featureData['CUST_CODE'] = featureData.CUSTID.astype('category').cat.codes
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from keras.utils import np_utils
```

```python
featureData['CUSTTYPE_CODE'] = featureData.CUSTTYPE.astype('category').cat.codes
```

```python
typemap = featureData[["CUSTTYPE_CODE","CUSTTYPE"]].drop_duplicates()
```

```python
c = featureData.CUSTTYPE.astype('category')
```

```python
d = dict(enumerate(c.cat.categories))
d
```

```python
featureData = featureData.drop(columns=["CUSTTYPE"])
```

```python
labelCol = ['CUSTTYPE_CODE']
labelCol
```

```python
featureCol = list(featureData.columns)
```

```python
featureCol.remove(labelCol[0])
```

```python
totalLen = len(featureData)
divStd = totalLen*0.8
divStd
```

```python
train_features = featureData[featureData.index < divStd][featureCol]
train_labels = featureData[featureData.index < divStd][labelCol]
test_features = featureData[featureData.index >= divStd][featureCol]
test_labels = featureData[featureData.index >= divStd][labelCol]
```

```python
outputCat = len(test_labels.drop_duplicates())
```

```python
from keras.models import Sequential
from keras.layers import Dense
#모델 생성
model = Sequential()

#get number of columns in training data
n_cols = train_features.shape[1]
n_cols
```

```python
#add layers to model
model.add(Dense(250, activation='relu', input_shape=(n_cols,)))
model.add(Dense(250, activation='relu'))
model.add(Dense(250, activation='relu'))
model.add(Dense(outputCat, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

### 모델 훈련

```python
from keras.callbacks import EarlyStopping
#set early stopping monitor so the model stops training when it won't improve anymore
early_stopping_monitor = EarlyStopping(patience=2000)
#train model
model.fit(train_features, train_labels, epochs=1000, validation_split=0.2, callbacks=[early_stopping_monitor])
```

```python
test_loss, test_acc = model.evaluate(test_features, test_labels)

print('테스트 정확도:', test_acc)
```
