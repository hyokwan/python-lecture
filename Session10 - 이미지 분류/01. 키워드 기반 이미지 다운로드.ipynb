```python
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys 
import time 
import os 
import urllib.request
```

```python
def createFolder(directory): 
    try: 
        if not os.path.exists(directory): 
            os.makedirs(directory) 
    except OSError: 
        print ('Error: Creating directory. ' + directory)
```

### 키워드 정의

```python
keyword='그랜저' 
```

```python
targetFolder = "../images/classImages/"+keyword
```

```python
createFolder(targetFolder)
```

### 크롬브라우저 기동

```python
chromedriver = "../externlib/chromedriver/chromedriver.exe"

driver = webdriver.Chrome(chromedriver) 
driver.implicitly_wait(3)
```

```python
driver.get('https://www.google.co.kr/imghp?hl=ko') 
Keyword=driver.find_element_by_xpath('//*[@id="sbtc"]/div/div[2]/input') 
Keyword.send_keys(keyword) 
Keyword.send_keys(Keys.RETURN)
```

```python
# //*[@id="islmp"]/div/div/div/div[2]/span
```

```python
print(keyword+' 스크롤 중 .............') 
elem = driver.find_element_by_tag_name("body") 
#결과더보기path
resultView = '//*[@id="islmp"]/div/div/div/div[1]/div[2]/div[2]/input'
for i in range(60): 
    elem.send_keys(Keys.PAGE_DOWN) 
    time.sleep(0.1) 
    try: 
        driver.find_element_by_xpath(resultView).click() 
        for i in range(60): 
            elem.send_keys(Keys.PAGE_DOWN) 
            time.sleep(0.1) 
    except: 
        pass
```

```python
#islrg > div.islrc > div.isv-r.PNCib.MSM1fd.BUooTd.fT6ABc > a.wXeWr.islib.nfEiy > div.bRMDJf.islir > img
```

```python
links=[]
# 썸네일 이미지 class name은 img.rg_i.Q4LuWd
# class명은 그대로 id명은 #
images = driver.find_elements_by_css_selector("img.rg_i.Q4LuWd") 
for image in images: 
    if image.get_attribute('src')!=None: 
        links.append(image.get_attribute('src')) 
print(keyword+' 찾은 이미지 개수:',len(links)) 
time.sleep(2)
```

```python
for k,i in enumerate(links): 
    url = i 
    start = time.time() 
    # url내용을 바로 저장 특정경로에 저장
    urllib.request.urlretrieve(url, targetFolder+"/"+keyword+str(k)+".jpg")  
    print(str(k+1)+'/'+str(len(links))+' '+keyword+' 다운로드 중....... Download time : '+str(time.time() - start)[:5]+' 초') 
    print(keyword+' ---다운로드 완료---') 
    
driver.close()
```
