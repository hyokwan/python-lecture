```python
nations = ["A01","미국","A02","한국","A03","프랑스"] 

i = 0

while i < len(nations):
    if(i%2!=0):
        print(nations[i])
    i = i+1
```

```python
nations = ["A01","미국","A02","한국","A03","프랑스"] 

for i in range (0,len(nations),2):
    print(nations[i])
```

```python
nations = ["A01","미국","A02","한국","A03","프랑스"] 

for i in nations:
    if(nations.index(i)%2!=0):
        print(i)
```
