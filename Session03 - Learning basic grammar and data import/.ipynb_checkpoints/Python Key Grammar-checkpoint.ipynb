```python
tvList = [ 'UN40EN001', 'UN40EN002', 'UN40EN003', 'UN40EN004']
```

## 2. 조건문

#### 따라하기: 모든 목록에 LEDTV를 붙여야한다. (단, 이미 붙어있으면 제외한다)

```python
tvList2 = [ 'UN40EN001', 'LEDTV_UN40EN002', 'LEDTV_UN40EN003', 'UN40EN004']
```

```python
preFix = "LEDTV_"
### range: last index is not included
### list(range(0,4))
for i in range(0,4):
    preFixCount = tvList2[i].count(preFix)
    if(preFixCount == 0):
        tvList2[i] = preFix + tvList2[i]
    else:
        tvList2[i] = tvList2[i]
```

```python
tvList2
```

#### 실습1: 모든 목록에 LEDTV를 붙여야한다. (단, 한번만 붙어야함)

```python
tvList = [ 'UN40EN001', 'LEDTV_UN40EN002', 'LEDTV_LEDTV_UN40EN003', 'UN40EN004']
```

```python
preFix = 'LEDTV_'
```

```python
i=0
while( i < len(tvList)):
    tvListCount = tvList[i].count(preFix)
    tvListCount

    ## 만약 count == 0 prefix 를 앞에 붙인다
    if(tvListCount == 0):
        tvList[i] = preFix + tvList[i]
    ## 만약 count == 1 현상태 유지
    elif(tvListCount == 1):
        tvList[i] = tvList[i]
    ## 만약 count > 1 prefix를 삭제 후 붙인다
    else:
        tvList[i] = tvList[i].replace(preFix, "")
        tvList[i] = preFix + tvList[i]
    i=i+1
```

```python
tvList
```

## 3. 함수사용

```python
# This function adds two numbers 
def roundTwo(value):
   step1 = value * 100
   step2 = round(step1)
   step3 = step2/100
   return step3

a = 10
b = 5.3
display2Digit = a/b
print(display2Digit)
answer = roundTwo(display2Digit)
answer
```

#### 실습. 리스트를 인수로 받아서 최대, 최소값을 뺀 평균을 구하는 함수를 생성하세요.

```python
def listAverage(inputList):
    inputList.sort()
    del inputList[len(inputList)-1]
    del inputList[0]

    outPut = sum(inputList)/len(inputList)
    return outPut
```

```python
testList = [1,2,33,4,5,6,7,222]
answer = listAverage(testList)
answer
```
